graphical representation of city by number of matches 

```python
sns.countplot(y='city', data=data)
plt.title('venue of matches\n')
plt.xlabel('\nNo. of matches held')
plt.ylabel('Cities\n')
plt.xlim([0,120])
plt.show()
```

from this graph we got that maximum number of matches is played in mumbai

counting the number of wins by each team

```python
data['winner'].value_counts()
```

Graphical representation of same team by number of wins 

```python
df=data['winner'].value_counts()
fig, ax=plt.subplots()
ax.set_xlim([0,120])
sns.barplot(y=df.index, x=df, orient='h')
plt.title('Number of matches won by each team\n')
plt.xlabel("\nNumber of matches won")
plt.ylabel('Teams\n')
plt.show()
```

we got that mumbai had won most of the matches in ipl

graphical representation of number of matches in each season 

```python
sns.countplot(x='season', data=data)
plt.title('Number of matches held every season\n')
plt.xlabel('\nSeason')
plt.ylabel('Number of matches held\n')
plt.show()
```

maximum number of matches played in season 2013

Finding out the players who had won most number of man of the matches 

```python
man_of_matches=data['player_of_match'].value_counts()[:10]
man_of_matches
```

graphical representation of the same

```python
plt.rcParams['figure.figsize'] = (20, 8)
fig, ax=plt.subplots()
ax.set_ylim([0,25])
ax.set_title('Top 10 players based on number of Man of matches awards won\n')
sns.barplot(x=man_of_matches.index, y=man_of_matches, orient='v')
plt.ylabel('Number of Man of matches awards won\n')
plt.xlabel('\nPlayers')
plt.show()
```
